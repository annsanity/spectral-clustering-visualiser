```python
df_export_import = df_export_import[df_export_import["HDI Rank"] != ".."]
df_foreign_inflow = df_foreign_inflow[df_foreign_inflow["HDI Rank"] != ".."]
df_dev_assistance = df_dev_assistance[df_dev_assistance["HDI Rank"] != ".."]
df_priv_capital_flow = df_priv_capital_flow[df_priv_capital_flow["HDI Rank"] != ".."]
df_remit_inflow = df_remit_inflow[df_remit_inflow["HDI Rank"] != ".."]
```

Remove duplicate data

```python
df_export_import = df_export_import.drop_duplicates(subset= "Country")
df_foreign_inflow = df_foreign_inflow.drop_duplicates(subset= "Country")
df_dev_assistance = df_dev_assistance.drop_duplicates(subset= "Country")
df_priv_capital_flow = df_priv_capital_flow.drop_duplicates(subset= "Country")
df_remit_inflow = df_remit_inflow.drop_duplicates(subset= "Country")
```

Replacing ".." with NaN

```python
# Replace '..' with np.nan
df_export_import.replace('..', np.nan, inplace=True)
df_foreign_inflow.replace('..', np.nan, inplace=True)
df_dev_assistance.replace('..', np.nan, inplace=True)
df_priv_capital_flow.replace('..', np.nan, inplace=True)
df_remit_inflow.replace('..', np.nan, inplace=True)
```

Applying LOCF imputation technique for missing for the year 2019

```python
df_export_import.iloc[:, 2:] = df_export_import.iloc[:,2:].ffill(axis=1)
df_foreign_inflow.iloc[:, 2:] = df_foreign_inflow.iloc[:,2:].ffill(axis=1)
df_dev_assistance.iloc[:, 2:] = df_dev_assistance.iloc[:,2:].ffill(axis=1)
df_priv_capital_flow.iloc[:, 2:] = df_priv_capital_flow.iloc[:,2:].ffill(axis=1)
df_remit_inflow.iloc[:, 2:] = df_remit_inflow.iloc[:,2:].ffill(axis=1)
```

```python
# Keeping desired columns that is 2019 (latest) data
df_export_import = df_export_import[["HDI Rank","Country","2019"]]
df_foreign_inflow = df_foreign_inflow[["HDI Rank","Country","2019"]]
df_dev_assistance = df_dev_assistance[["HDI Rank","Country","2018"]]
df_priv_capital_flow = df_priv_capital_flow[["HDI Rank","Country","2019"]]
df_remit_inflow = df_remit_inflow[["HDI Rank","Country","2019"]]
```

```python
# Renaming the latest year column to indicate the respective dataframe's name
df_export_import.rename(columns={'2019': 'Export_Import'}, inplace=True)
df_foreign_inflow.rename(columns={'2019': 'Foreign_Inflow'}, inplace=True)
df_dev_assistance.rename(columns={'2018': 'Dev_Assistance'}, inplace=True)
df_priv_capital_flow.rename(columns={'2019': 'Priv_Capital_Flow'}, inplace=True)
df_remit_inflow.rename(columns={'2019': 'Remittance_Inflow'}, inplace=True)
```

```python
# Merging the dataframes on 'HDI Rank' and 'Country'
df = df_export_import
df = df.merge(df_foreign_inflow, on=['HDI Rank', 'Country'], how='outer')
df = df.merge(df_dev_assistance, on=['HDI Rank', 'Country'], how='outer')
df = df.merge(df_priv_capital_flow, on=['HDI Rank', 'Country'], how='outer')
df = df.merge(df_remit_inflow, on=['HDI Rank', 'Country'], how='outer')
```

```python
# Re-arranging the columns
df = df.reindex(columns=['Country', 'HDI Rank', 'Export_Import','Foreign_Inflow','Dev_Assistance','Priv_Capital_Flow','Remittance_Inflow'])
df.head(10)
```

```python
#Normalizing the data and making all columns to be in similar range 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['HDI Rank', 'Export_Import','Foreign_Inflow','Dev_Assistance','Priv_Capital_Flow','Remittance_Inflow']] = scaler.fit_transform(df[['HDI Rank', 'Export_Import','Foreign_Inflow','Dev_Assistance','Priv_Capital_Flow','Remittance_Inflow']])
```

```python
from missforest import MissForest

# Initialize the magical forest
imputer = MissForest()

# Impute away
df_imputed = imputer.fit_transform(df.iloc[:,1:])
```

```python
df = df.iloc[:,0:1]
df = df.join(df_imputed)
```

```python
df = df.dropna()
```
